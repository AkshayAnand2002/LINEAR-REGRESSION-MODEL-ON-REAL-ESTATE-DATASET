## Linear Regression

```python
#Linear Regression

from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
#Simple Linear Regression

X2=real_estate[["X2 house age"]]
Y=real_estate[["Y house price of unit area"]]
```

```python
lm.fit(X2,Y) #MODEL TRAINS AND RUNS FROM THE DATA.
Yhat=lm.predict(X2)
Yhat
```

```python
print("Intercept=",lm.intercept_)
print("Coefficient=",lm.coef_)
print("R squared=", lm.score(X2,Y))
#Y=BO + B1(X)
#B0-INTERCEPT AND B1-COEFF OR SLOPE I.E. lm.coef_
```

```python
#residual plot

sns.residplot(x="X2 house age", y="Y house price of unit area", data=real_estate)
```

```python
#Actual values vs Predicted values of house price per unit area

ax1=sns.histplot(real_estate["Y house price of unit area"],color="r", label="Actual Value",kde=True, stat="density", linewidth=0)
sns.histplot(Yhat, color="b", label="Fitted Values", ax=ax1,kde=True, stat="density", linewidth=0)
```

```python
#blue color represents fitted value and red color represents actual values which are significantly different.
```

```python
#so instead of using only one column for training we use more columns.
```

```python
Z=real_estate[["X1 transaction date","X2 house age", "X3 distance to the nearest MRT station","X4 number of convenience stores","X5 latitude","X6 longitude"]]
lm.fit(Z,Y) #training.
Yhat1=lm.predict(Z)
Yhat1
```

## Ridge

```python
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV
np.linspace(1,100,100)
```

```python
np.arange(1,10,0.5)
```

```python
lambdas=np.linspace(1,100,100)
```

```python
params={'alpha':lambdas}
```

```python
model=Ridge(fit_intercept=True) #RIDGE REGRESSION
```

```python
grid_search=GridSearchCV(model,param_grid=params,cv=10,scoring='neg_mean_absolute_error')
#cv MEANS CROSS VALIDATIONS.
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.cv_results_
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
#COMPARES THE RESULT AND PROVIDES THE VALUES FOR WHICH THE TOP 3 OR BEST 3 MODELS WHICH HAVE PERFORMED THE BEST.
```

```python
report(grid_search.cv_results_,5) # ARGUEMENT 5 GIVES TOP 5 MODELS.
```

```python
ridge_model=grid_search.best_estimator_
```

```python
ridge_model.fit(x_train,y_train)
```

```python
list(zip(x_train.columns,ridge_model.coef_))
```

```python
#WE GET COEFFICENTS FOR CORRESPONDING COLUMNS BY RIDGE REGRESSION WHICH HAVE REDUCED AS COMPARED TO PREVIOUS VALUE.
```

## Lasso

```python
lambdas=np.linspace(1,10,100)
model=Lasso(fit_intercept=True)
params={'alpha':lambdas}
```

```python
grid_search=GridSearchCV(model,param_grid=params,cv=10,scoring='neg_mean_absolute_error')
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
report(grid_search.cv_results_,5)
```

```python
lasso_model=grid_search.best_estimator_
lasso_model.fit(x_train,y_train)
list(zip(x_train.columns,lasso_model.coef_))
```

```python
#HERE THE COEFFICIENTS HAVE BEEN ACTUALLY BROUGHT DOWN A LOT TO 0 (EXACTLY 0).
```
